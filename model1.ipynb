```python
data0 = pd.read_csv('../data/data.csv')
data = data_preprocess.pre_process_model(data0)
```

```python
from scipy.spatial.distance import euclidean, pdist, squareform
from sklearn.metrics import jaccard_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data[['longitude', 'latitude', 'event_month_ctr', 'transformed_fatalities', 'days_from_19970101']] = scaler.fit_transform(data[['longitude', 'latitude', 'event_month_ctr', 'transformed_fatalities', 'days_from_19970101']])
data
```

```python
def jaccob_dist(u, v):
    categorical_dist = (u[:4] != v[:4]).sum() / 4
    #return jaccard_score(u[:4], v[:4], average='samples')
    return categorical_dist

data_small = data[:1000]
dists = pdist(data_small.iloc[:,4:], metric='euclidean')
cont_euclidean_dist = pd.DataFrame(squareform(dists), columns=data_small.index, index=data_small.index)
cont_sim_mat = np.exp(-cont_euclidean_dist**2)

dists = pdist(data_small.iloc[:,:4], metric=jaccob_dist)
cat_jac_dist = pd.DataFrame(squareform(dists), columns=data_small.index, index=data_small.index)
cat_sim_mat = 1-cat_jac_dist

sim_mat = cont_sim_mat / 2 + cat_sim_mat / 2
```

```python
from sklearn.cluster import SpectralClustering

clustering = SpectralClustering(n_clusters=2,
         assign_labels='discretize',
         random_state=0, 
         n_components = 2,
         n_jobs=2).fit(sim_mat)
clustering.labels_
```
